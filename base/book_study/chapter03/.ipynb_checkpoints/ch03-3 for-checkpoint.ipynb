### 예제를 통한 for문 이해하기

```python
# 전형적인 for문
test_list = ['one', 'two', 'three']
for i in test_list:
    print(i)
```

```python
# 다양한 for문 사용
a = [(1,2), (3,4), (5,6)]  # 튜플 사용
for (first, last) in a:
    print(first + last)
```

```python
# for문의 응용
scores = [90, 25, 67, 45, 80]
number = 0
for s_score in scores:
    number = number+1
    if s_score >= 60:
        print("%d번 학생은 합격입니다." % number)
    else:
        print("%d번 학생은 불합격입니다." % number)
```

```python
scores = [90, 25, 67, 45, 80]
number = 0
for s_score in scores:
    number = number+1
    print("%d번 학생은 합격입니다." % number) if s_score >= 60 else print("%d번 학생은 불합격입니다." % number)
```

### for문과 continue문

```python
# marks2.py
marks = [90, 25, 67, 45, 80]
number = 0
for mark in marks:
    number += 1
    if mark < 60: continue
    print("%d번 학생 축하합니다. 합격입니다. " % number)
```

### for문과 range함수

```python
sum = 0
for i in range(1, 11):  # 1~10까지 누적시킴
    sum = sum + i

print(sum)
```

```python
# marks3.py
marks = [90, 25, 67, 45, 80]
for number in range(len(marks)):
    if marks[number] < 60: continue
    print("%d번 학생 축하합니다. 합격입니다." % (number+1))
```

```python
# for & range 사용해 구구단 만들기
for i in range(2, 10):
    for j in range(1, 10):
        print(i*j, end=" ") # 해당 결과값을 그 줄에 계속해서 출력
    print('') # 결과값을 다음 줄부터 출력
```

### 리스트 내포(List comprehension)

```python
a = [1,2,3,4]
result = []
for num in a:
    result.append(num*3)

print(result)
```

```python
# 위와 같은 코드
a = [1,2,3,4]
result = [num * 3 for num in a]
print(result)
```

```python
# 리스트 내포 안에 if 조건문 사용
result = [num * 3 for num in a if num % 2 == 0]
print(result)
```

```python
# 구구단
result = [x*y for x in range(2,10)
         for y in range(1,10)]
print(result)
```

### p.147 Q6 연습문제

```python
numbers = [1,2,3,4,5]
result = []
for n in numbers:
    if n%2 == 1:
        result.append(n*2)
result
```

```python
# 리스트 내포 표현
numbers = [1,2,3,4,5]
result = [n*2 for n in numbers if n%2 == 1]
print(result)
```
